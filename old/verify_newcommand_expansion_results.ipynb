```python
import re
import pandas as pd 
#output is obtained by running ssdeep on the pdfs generated on the rawlatex file and the pdf generated by the texfile with the newcommand expanded

with open('output','r') as f:
    data = [x.strip() for x in f.readlines()]

df = pd.read_csv('output',sep=',')
df = df[df['filename'].apply(lambda x: True if not re.findall('filename',str(x)) else False)]
del df['filename']
df = df[df.ssdeep.apply(lambda x: True if len(str(x))>10 else False)]
df.ssdeep = df.ssdeep.apply(lambda x: str(x).split(':')[1].strip())
df.sort_values('ssdeep',inplace=True)
df.reset_index(inplace=True,drop=True)


def find_similar(index):
    #if the hash matches up to the last char the difference is a few characters of text related to how pdflatex is processing the tex files, but the generated output will be equivalent.
    indicies = [x for x in df.index.to_list() if x!=index]
    resp = df[df.ssdeep.apply(lambda x:str(x)[:37] == df.ssdeep.loc[index][:37])]['1.1--blocksize:hash:hash'].to_list()
    if len(resp) >= 2:
        return resp
    else:
        return False
```

```python
from time import sleep
results = set()
for ix in df.index.to_list():
    resp = find_similar(ix)
    if resp:
        resp = [x.replace('latex-in-arxiv/2003/PDFS/','',).replace('_newcmd_expanded','') for x in find_similar(ix)]
        if resp:
            if len(resp) == 2:
                if resp[0] == resp[1]:
                    results.add(resp[0])
```
